```python
import os
from netCDF4 import Dataset
file_path = r'F:\Academics\Research_paper\Programs\Data\spr\gmi\training\on_swath\spatial\gmi'

for i in os.listdir(file_path):

    if i.startswith('gmi') and i.endswith('.nc'):
        path = os.path.join(file_path, i)
        
        db = Dataset(path, 'r')
        
        var = db.variables.keys()

        print(f"\nVar in {i}:")
        for var1 in var:
            print(var1)

db.close()
```
